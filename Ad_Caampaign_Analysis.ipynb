```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from the provided data
data = pd.read_csv('Data from marketing team - Raw Data.csv')
```

```python
# Display basic information about the data
print(data.info())
```

```python
# Explore the first few rows of the dataset
print(data.head())
```

```python
# Check for any missing values
print(data.isnull().sum())
```

```python
# Summary statistics for numeric columns
print(data.describe())
```

 Bar Chart - Comparing Click-Through Rate (CTR) by Campaign Name


```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Campaign Name', y='Click-Through Rate (CTR)', data=data)
plt.title('Click-Through Rate (CTR) by Campaign Name')
plt.xlabel('Campaign Name')
plt.ylabel('Click-Through Rate (CTR)')
plt.xticks(rotation=90)
plt.show()
```

 Bar Chart - Comparing Cost Per Click (CPC) by Campaign Name

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Campaign Name', y='Cost Per Click (CPC)', data=data)
plt.title('Cost Per Click (CPC) by Campaign Name')
plt.xlabel('Campaign Name')
plt.ylabel('Cost Per Click (CPC)')
plt.xticks(rotation=90)
plt.show()
```

Least Click-Through Rate (CTR)

```python
lowest_ctr_campaign = data.loc[data['Click-Through Rate (CTR)'].idxmin()]
print("Least CTR Campaign:\n", lowest_ctr_campaign)
```

Highest Cost Per Click (CPC)

```python
highest_cpc_campaign = data.loc[data['Cost Per Click (CPC)'].idxmax()]
print("Highest CPC Campaign:\n", highest_cpc_campaign)
```

Bar chart for - Metrics: Click-Through Rate (CTR), Unique Click-Through Rate (Unique CTR), Cost Per Click (CPC), Cost per Result (CPR)

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Campaign Name', y='Cost per Result (CPR)', data=data)
plt.title('Cost per Result (CPR) by Campaign Name')
plt.xlabel('Campaign Name')
plt.ylabel('Cost per Result (CPR)')
plt.xticks(rotation=90)
plt.show()
```

Scatter Plot:

Metrics: Reach vs. Click-Through Rate (CTR), Unique Clicks vs. Impressions

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Reach', y='Click-Through Rate (CTR)', data=data)
plt.title('Reach vs. Click-Through Rate (CTR)')
plt.xlabel('Reach')
plt.ylabel('Click-Through Rate (CTR)')
plt.show()
```

Pie Chart:

Metrics: Composition of Audience (count of each category)

```python
plt.figure(figsize=(8, 8))
data['Audience'].value_counts().plot(kind='pie', autopct='%1.1f%%', explode=(0.1, 0), colors=['lightblue', 'lightgreen'])
plt.title('Composition of Audience')
plt.ylabel('')
plt.show()
```

Stacked Bar Chart:

Metrics: Comparing the composition of multiple data series (e.g., Age breakdown between campaigns)

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Campaign Name', y='Impressions', hue='Age', data=data)
plt.title('Impressions by Age and Campaign')
plt.xlabel('Campaign Name')
plt.ylabel('Impressions')
plt.xticks(rotation=90)
plt.show()
```

Heatmap:

Metrics: Correlation matrix between different numerical variables (e.g., Reach, Impressions, Clicks, etc.)

```python
plt.figure(figsize=(10, 6))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

Histogram:

Metrics: Distribution of a numerical variable (e.g., Reach, Impressions, etc.)

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Reach'], bins=20, kde=True)
plt.title('Distribution of Reach')
plt.xlabel('Reach')
plt.ylabel('Count')
plt.show()
```
